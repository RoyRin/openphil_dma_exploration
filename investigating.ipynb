```python
import pandas as pd
import numpy as np
```

```python
path = "Release20_mapdata.csv"
```

```python
df = pd.read_csv(path)
```

```python
def dollars_to_int(df, name):
    if (df[name].dtype == np.int64):
        return
    df[name]=  df[name].apply(lambda x: "".join(x.lstrip()[1:].split(','))).astype(np.int64)

def number_to_int(df, name):
    if (df[name].dtype == np.int64):
        return
    df[name]=  df[name].apply(lambda x: "".join(x.lstrip().split(','))).astype(np.int64)
    
dollars_to_int(df, 'Est. Cost 9Apr20 to 25Oct29') 
number_to_int(df, "Total Airings 9Apr20 to 25Oct20")
dollars_to_int(df, 'Est. Cost 12Oct20 to 25Oct20')
number_to_int(df, "Total Airings 12Oct20 to 25Oct20")
```

```python
df['average_cost_per_airing 9Apr20 to 25Oct20'] =  (df['Est. Cost 9Apr20 to 25Oct29'])/ df['Total Airings 9Apr20 to 25Oct20']
df['average_cost_per_airing 12Oct20 to 25Oct20'] = df['Est. Cost 12Oct20 to 25Oct20']/ df['Total Airings 12Oct20 to 25Oct20']
```
